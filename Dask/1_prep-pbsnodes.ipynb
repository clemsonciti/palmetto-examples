```python
import json
import pandas as pd
import numpy as np
```

```python
with open('/project/rcde/xdmod_analysis/pbsnodes.csv', 'r') as f: 
    pbsnodes = json.load(f)
```

```python
# top-level keys
[k for k, _ in pbsnodes.items()]
```

```python
df = pd.DataFrame(pbsnodes['nodes']).T
df
```

```python
# extract the resources column as a dataframe and join onto data
df_res = pd.DataFrame(df.resources_available.tolist(), index = df.index)
df_res
```

```python
df = df.join(df_res)
```

```python
# drop irrelevant cols
df = df.drop(['state', 'resources_assigned', 'resources_available', 'resv_enable', 'last_state_change_time', 'last_used_time', 'jobs'], axis=1)
```

```python
df
```

```python
# drop missing pcpus
df = df.dropna(subset=['pcpus'])
```

```python
# add host column to match job data
df = df.reset_index().rename(columns={'index': 'hostname'})
```

### Detailed gpu info

```python
df.gpu_model = df.gpu_model.apply(lambda x: str(x).split(',')[0])
```

```python
gpu_map = {k: k for k in df.gpu_model.unique()}
gpu_map['dgxa100'] = 'a100'
gpu_map['dgx1'] = 'v100'
gpu_map
```

```python
df.gpu_model = df.gpu_model.map(gpu_map)
```

```python
# save
df.to_csv('/project/rcde/xdmod_analysis/host_info.csv', index=None)
```
